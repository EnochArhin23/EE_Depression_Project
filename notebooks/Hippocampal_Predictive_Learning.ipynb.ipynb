```python
import numpy as np
import matplotlib.pyplot as plt

from ratinabox.Environment import Environment
from ratinabox.Agent import Agent
from ratinabox.Neurons import PlaceCells

# --- Create enriched environment ---
env = Environment(params={
    "boundary": [[0,0],[1,0],[1,1],[0,1]],
    "walls": [
        [[0.3,0.3],[0.7,0.3]],
        [[0.3,0.7],[0.7,0.7]],
        [[0.3,0.3],[0.3,0.7]],
        [[0.7,0.3],[0.7,0.7]],
    ]
})

agent = Agent(env)
place_cells = PlaceCells(agent, params={"n": 60})

# --- Predictive neuron weights ---
n_cells = 60
weights = np.random.randn(n_cells) * 0.01
learning_rate = 0.005  # This is what we will vary for depression vs enrichment

# Store prediction error over time
errors = []

# --- Simulation ---
T = 300
delay_steps = int(1.0 / agent.dt)  # 1 second prediction window
future_positions = []

for t in range(int(T / agent.dt)):
    agent.update()
    place_cells.update()

    # Current place cell activity
    pc_activity = place_cells.firingrate

    # Save future positions
    future_positions.append(agent.pos.copy())

    if t > delay_steps:
        target_pos = future_positions[t - delay_steps]

        # Predict x-position only (simplify)
        prediction = np.dot(weights, pc_activity)

        error = target_pos[0] - prediction
        errors.append(error)

        # BTSP-like learning rule
        weights += learning_rate * error * pc_activity

print("✅ Predictive learning simulation finished")

# --- Plot learning curve ---
plt.figure(figsize=(8,4))
plt.plot(np.abs(errors))
plt.title("Prediction Error Over Time (Learning)")
plt.xlabel("Time step")
plt.ylabel("Absolute Error")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from ratinabox.Environment import Environment
from ratinabox.Agent import Agent
from ratinabox.Neurons import PlaceCells


def run_experiment(env_type="standard", learning_rate=0.005):
    # --- Environment setup ---
    if env_type == "enriched":
        env = Environment(params={
            "boundary": [[0,0],[1,0],[1,1],[0,1]],
            "walls": [
                [[0.3,0.3],[0.7,0.3]],
                [[0.3,0.7],[0.7,0.7]],
                [[0.3,0.3],[0.3,0.7]],
                [[0.7,0.3],[0.7,0.7]],
            ]
        })
    else:
        env = Environment()

    agent = Agent(env)
    place_cells = PlaceCells(agent, params={"n": 60})

    # Predictive neuron
    weights = np.random.randn(60) * 0.01
    errors = []

    T = 300
    delay_steps = int(1.0 / agent.dt)
    future_positions = []

    for t in range(int(T / agent.dt)):
        agent.update()
        place_cells.update()

        pc_activity = place_cells.firingrate
        future_positions.append(agent.pos.copy())

        if t > delay_steps:
            target_pos = future_positions[t - delay_steps][0]
            prediction = np.dot(weights, pc_activity)

            error = target_pos - prediction
            errors.append(abs(error))

            weights += learning_rate * error * pc_activity

    return errors


# --- Run 3 brains ---
errors_depressed = run_experiment("standard", learning_rate=0.001)
errors_normal = run_experiment("standard", learning_rate=0.005)
errors_enriched = run_experiment("enriched", learning_rate=0.02)


# --- Plot comparison ---
plt.figure(figsize=(9,5))
plt.plot(errors_depressed, label="Depression (low plasticity)")
plt.plot(errors_normal, label="Normal")
plt.plot(errors_enriched, label="Enriched (high plasticity)")
plt.xlabel("Time step")
plt.ylabel("Prediction Error")
plt.title("Environmental Enrichment Improves Predictive Learning")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from ratinabox.Environment import Environment
from ratinabox.Agent import Agent
from ratinabox.Neurons import PlaceCells


def spatial_information(rate_map):
    r = rate_map.flatten()
    r = r / (np.sum(r) + 1e-12)
    mean_r = np.mean(r)
    info = np.nansum(r * np.log2((r + 1e-12) / (mean_r + 1e-12)))
    return info


def measure_placecell_quality(env_type="standard"):
    # --- Environment ---
    if env_type == "enriched":
        env = Environment(params={
            "boundary": [[0,0],[1,0],[1,1],[0,1]],
            "walls": [
                [[0.3,0.3],[0.7,0.3]],
                [[0.3,0.7],[0.7,0.7]],
                [[0.3,0.3],[0.3,0.7]],
                [[0.7,0.3],[0.7,0.7]],
            ]
        })
    else:
        env = Environment()

    agent = Agent(env)
    place_cells = PlaceCells(agent, params={"n": 60})

    # --- Record data ---
    positions = []
    activities = []

    for _ in range(int(300 / agent.dt)):
        agent.update()
        place_cells.update()
        positions.append(agent.pos.copy())
        activities.append(place_cells.firingrate.copy())

    positions = np.array(positions)
    activities = np.array(activities)

    # --- Create spatial bins ---
    bins = 20
    x_edges = np.linspace(0, 1, bins+1)
    y_edges = np.linspace(0, 1, bins+1)

    infos = []

    for cell in range(60):
        rate_map = np.zeros((bins, bins))
        occupancy = np.zeros((bins, bins))

        for t in range(len(positions)):
            x, y = positions[t]
            xi = np.digitize(x, x_edges) - 1
            yi = np.digitize(y, y_edges) - 1

            if 0 <= xi < bins and 0 <= yi < bins:
                rate_map[xi, yi] += activities[t, cell]
                occupancy[xi, yi] += 1

        # Avoid divide by zero
        rate_map = rate_map / (occupancy + 1e-12)

        infos.append(spatial_information(rate_map))

    return infos


# --- Run comparison ---
info_standard = measure_placecell_quality("standard")
info_enriched = measure_placecell_quality("enriched")

# --- Plot ---
plt.figure(figsize=(8,5))
plt.boxplot([info_standard, info_enriched], labels=["Standard", "Enriched"])
plt.ylabel("Spatial Information Content")
plt.title("Environmental Enrichment Improves Place Cell Information")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from ratinabox.Environment import Environment
from ratinabox.Agent import Agent
from ratinabox.Neurons import PlaceCells


def run_dendritic_model(env_type="standard", learning_rate=0.01):
    # --- Environment ---
    if env_type == "enriched":
        env = Environment(params={
            "boundary": [[0,0],[1,0],[1,1],[0,1]],
            "walls": [
                [[0.3,0.3],[0.7,0.3]],
                [[0.3,0.7],[0.7,0.7]],
                [[0.3,0.3],[0.3,0.7]],
                [[0.7,0.3],[0.7,0.7]],
            ]
        })
    else:
        env = Environment()

    agent = Agent(env)
    place_cells = PlaceCells(agent, params={"n": 60})

    # Two weight branches (dendrite + soma)
    w_fast = np.random.randn(60) * 0.01
    w_slow = np.random.randn(60) * 0.01

    errors = []

    T = 300
    delay_steps = int(1.0 / agent.dt)        # predict 1 sec ahead
    memory_steps = int(0.5 / agent.dt)      # dendritic memory (0.5 sec)

    history = []
    future_positions = []

    for t in range(int(T / agent.dt)):
        agent.update()
        place_cells.update()

        pc = place_cells.firingrate.copy()
        history.append(pc)
        future_positions.append(agent.pos.copy())

        if t > delay_steps and t > memory_steps:
            target_x = future_positions[t - delay_steps][0]

            pc_now = history[t]
            pc_past = history[t - memory_steps]

            # Dendritic integration
            prediction = np.dot(w_fast, pc_now) + np.dot(w_slow, pc_past)

            error = target_x - prediction
            errors.append(abs(error))

            # Learn on both branches
            w_fast += learning_rate * error * pc_now
            w_slow += learning_rate * error * pc_past

    return errors


# --- Compare without vs with dendritic integration ---
errors_simple = run_dendritic_model("enriched", learning_rate=0.005)
errors_dendritic = run_dendritic_model("enriched", learning_rate=0.02)

plt.figure(figsize=(9,5))
plt.plot(errors_simple, label="Single-timescale learning")
plt.plot(errors_dendritic, label="Dendritic multi-timescale learning")
plt.xlabel("Time step")
plt.ylabel("Prediction Error")
plt.title("Dendritic Integration Improves Predictive Learning")
plt.legend()
plt.show()
```
